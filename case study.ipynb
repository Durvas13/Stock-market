```python
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
stock_symbol = "RELIANCE.NS"
start_data = '2024-01-01'
end_data = '2025-11-17'

data = yf.download(stock_symbol,start_data,end_data)

data.to_csv("stock.csv")
```

```python
data.head()
```

```python
# view summery
data.describe()
```

```python
print(data.isnull().sum())
```

```python
plt.Figure(figsize=(12,6))
plt.plot(data['Close'], label='closing price',color='blue')
plt.title("RELIANCE.NS closing price")
plt.xlabel("data")
plt.ylabel("price(INR)")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.Figure(figsize=(12,6))
plt.plot(data['Open'], label='open price',color='blue')
plt.plot(data['Close'], label='closeing price',color='green')
plt.title("RELIANCE.NS closing price")
plt.xlabel("data")
plt.ylabel("price(INR)")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.Figure(figsize=(12,6))
plt.plot(data['Low'], label='closing price',color='blue')
plt.plot(data['High'], label='closeing price',color='green')
plt.title("RELIANCE.NS closing price")
plt.xlabel("data")
plt.ylabel("price(INR)")
plt.legend()
plt.grid()
plt.show()
```

```python
data['MA50'] = data['Close'].rolling(window=50).mean()
data['MA200'] = data['Close'].rolling(window=200).mean()
```

```python
data.tail()
```

```python
plt.Figure(figsize=(12,6))
plt.plot(data['Close'], label='closing price',color='blue')
plt.plot(data['MA50'], label='closeing price',color='green')
plt.plot(data['MA200'], label='closeing price',color='pink')
plt.title("RELIANCE.NS closing price")
plt.xlabel("data")
plt.ylabel("price(INR)")
plt.legend()
plt.grid(True)
plt.show()
```
